```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x = 'fare', y = 'age', data = titanic)

# Most passengers paid less than $50 for a fare ticket, though a few passengers paid over $200
```

```python
sns.displot(titanic['fare'], kde = False, color = 'red', bins = 30)
plt.xlim([0,600])
```

```python
sns.boxplot(x = 'class', y = 'age', data = titanic, palette = 'rainbow')
plt.ylim([0,80])

# First class passengers tend to be older than second class (On Average)
# Second class passengers tend to be older than third class (On Average)
# Observation: The older and more established the passengers are, the more money they tend to have
```

```python
sns.swarmplot(x = 'class', y = 'age', data = titanic, palette= 'Set2',size = 3.5)

# There are a lot more third class passengers, with the majority ranging 18-30 years of age
```

```python
sns.countplot(x = 'sex', data = titanic)

# There are more male passengers than female
```

```python
sns.heatmap(titanic.corr(), cmap = 'coolwarm', annot= True)
plt.title('titanic.corr()')

# Not many adult males survived (Shown from the low correlation 'between adult_male' and 'survived' columns)
# This shows how women and children were more likely to be survivors
```

```python
g = sns.FacetGrid(data = titanic, col = 'sex')
g.map(sns.histplot, 'age')

# The distribution of age is similar among male and female passengers
```
